```python
data_list = [
    {
        'brand': '農純鄉',
        'link': link1,
        'xpath': xpath1,
        'cnt': 60,
        'ml': 100,
    },
    {
        'brand': '農純鄉',
        'link': link2,
        'xpath': xpath2,
        'cnt': 28,
        'ml': 100,
    },
        {
        'brand': '純煉紅棗枸杞',
        'link': link3,
        'xpath': xpath3,
        'cnt': 65,
        'ml': 60,
    },
        {
        'brand': '享食尚',
        'link': link7,
        'xpath': xpath7,
        'cnt': 27,
        'ml': 45,
    },
        {
        'brand': '老協珍',
        'link': link8,
        'xpath': xpath8,
        'cnt': 25,
        'ml': 42,
    },
        {
        'brand': '享食尚',
        'link': link9,
        'xpath': xpath9,
        'cnt': 60,
        'ml': 45,
    },
        {
        'brand': '享食尚',
        'link': link10,
        'xpath': xpath10,
        'cnt': 60,
        'ml': 45,
    }
]

result_df = pd.DataFrame()

for data_dict in data_list:
    link = data_dict['link']
    xpath = data_dict['xpath']
    
    driver = webdriver.Chrome()
    driver.get(link)
    
    price = driver.find_element("xpath", xpath).text
    
    price = price.replace('NT', '').replace('$', '').replace(',', '')
    price = int(price)
    
    data_dict['price'] = price
    data_dict['price2'] = (data_dict['cnt'] * data_dict['ml']) / price
    
    iteration_df = pd.DataFrame([data_dict])
    
    result_df = pd.concat([result_df, iteration_df], ignore_index=True)

dropCols=result_df.drop(columns='xpath')
dropCols
```

```python
dropCols=dropCols[['brand', 'price', 'cnt', 'ml', 'link', 'price2']]
dropCols

driver.get(link4) 
driver.find_element("xpath",xpath4).click()
price4=driver.find_element("xpath",xpath42).text
driver.quit()
data = {
    'brand': ['芳茲'],
    'price': [price4],
    'cnt': [50],
    'ml': [50],
    'link': [link4]
}

df4 = pd.DataFrame(data)
df4.price=df4.price.str.replace('NT','')
df4.price=df4.price.str.replace('$','')
df4.price=df4.price.str.replace(',','')
df4.price=df4.price.astype(int)

df4['price2'] = (df4.cnt * df4.ml) / df4.price
df4
```

```python
dfConcat=pd.concat([dropCols,df4])
dfConcat
```

```python
dfConcat['date'] = date1
dfConcat['date'] = dfConcat['date'].astype(str)
dfConcat
```

```python
csv1 = r'price.csv'
dfConcat.to_csv(csv1, mode='a',index=False,encoding='utf-8-sig')
dfNew = pd.read_csv(csv1)
```

```python
dfNew = dfNew[dfNew.brand != 'brand']
dfNew = dfNew.drop_duplicates()
dfNew.columns
dfNew = dfNew[['brand', 'price', 'cnt', 'ml', 'price2', 'date', 'link']]
dfNew = dfNew.sort_values(by=['date','price2'],ascending=[0,0])
dfNew['price2'] = dfNew['price2'].astype(float).round(2)

dfNew = dfNew.rename(columns={'brand': '品牌',
                        'price': '價格',
                        'cnt': '數量',
                        'ml': '每包ml',
                        'price2': '每元可買ml',
                        'date': '日期',
                        'link':'連結'
                        })

dfNew.to_csv(r'price2.csv',encoding='utf-8-sig',index=0)

dfNew
```
